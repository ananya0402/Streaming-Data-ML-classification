```python
# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
import csv


def read_csv(fileName):
    '''Read the CSV file with proper datatypes for each column'''
    data = []

    with open("Streaming_Linux_process.csv") as f:
        reader = csv.DictReader(f)
        for row in reader:
            data.append({'sequence':int(row['sequence']),'machine':int(row['machine']), 'PID':int(row['PID']),
                        'TRUN':int(row['TRUN']), 'TSLPI':int(row['TSLPI']), 'TSLPU':int(row['TSLPU']),
                        'POLI':str(row['POLI']), 'NICE':int(row['NICE']), 'PRI':int(row['PRI']),
                        'RTPR':int(row['RTPR']), 'CPUNR':int(row['CPUNR']), 'Status':str(row['Status']),
                        'EXC':int(row['EXC']), 'State':str(row['State']), 'CPU':float(row['CPU']), 'CMD': str(row['CMD'])})

    
    return data

def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'Streaming_Linux_process'
    cRows = list(read_csv('Streaming_Linux_process.csv'))
    
    print('Publishing records..')
    producer = connect_kafka_producer()
    
    '''Initialise empty lists for storing the records of each machine separately'''   
    mac_4=[]
    mac_5=[]
    mac_6=[]
    mac_7=[]
    mac_8=[]
    for i in range(len(cRows)):
  
        if(cRows[i]["machine"] == 4):
            mac_4.append(cRows[i])
        elif(cRows[i]["machine"] == 5):
            mac_5.append(cRows[i])
        elif(cRows[i]["machine"] == 6):
            mac_6.append(cRows[i])
        elif(cRows[i]["machine"] == 7):
            mac_7.append(cRows[i])
        if(cRows[i]["machine"] == 8):
            mac_8.append(cRows[i])
    
   
    max_records1 = 0 #stores maximum index for machine no 1
    min_records1= 0
    max_records2 = 0 #stores maximum index for machine no 2
    min_records2 = 0
    max_records3 = 0 #stores maximum index for machine no 3
    min_records3 = 0
    max_records4 = 0 #stores maximum index for machine no 4
    min_records4 = 0
    max_records5 = 0 #stores maximum index for machine no 5
    min_records5 = 0
   
    
    mac_4_1=[]
    mac_5_1=[]
    mac_6_1=[]
    mac_7_1=[]
    mac_8_1=[]
   
    #A while True loop to infinitely loop through records
    while True: 
        
        '''MACHINE NO.1'''
        #include timestamp as a key value pair for each record
        for value in mac_4:

            time_stamp =  int(dt.datetime.now().timestamp())
            value['ts']=time_stamp

            mac_4_1.append(value)
        
         #A random function to get random number of rows between (10,50)
        num1 = random.randint(10,50)
        max_records1 = num1
        l1=mac_4_1[min_records1:(min_records1+max_records1)]#records to be published
        

    #Create the data object to publish with the selected rows
        print("MachineID: 4")
        print("No. of records", num1)
        print("Records from: ",min_records1,':',min_records1+max_records1)
        publish_message(producer, topic, l1)
        min_records1 += num1
   
        if max_records1 >= len(mac_4):
            max_records1 = 0
            min_records1 = 0

        '''MACHINE NO.2'''
        #include timestamp as a key value pair for each record
        for value in mac_5:

            time_stamp =  int(dt.datetime.now().timestamp())
            value['ts']=time_stamp

            mac_5_1.append(value)
            
        

        
         #A random function to get random number of rows between (10,50)
        num2 = random.randint(10,50)
        max_records2 = num2
        l2 = mac_5_1[min_records2:(min_records2+max_records2)]#records to be published
        
    
    #Create the data object to publish with the selected rows
        print("MachineID: 5")
        print("No. of records", num2)
        print("Records from: ",min_records2,':',min_records2+max_records2)
        publish_message(producer, topic, l2)
        min_records2 += num2

        if max_records2 >= len(mac_5):
            max_records2 = 0
            min_records2 = 0
            
        '''MACHINE NO.3''' 
        #include timestamp as a key value pair for each record
        for value in mac_6:

            time_stamp =  int(dt.datetime.now().timestamp())
            value['ts']=time_stamp

            mac_6_1.append(value)
         #A random function to get random number of rows between (10,50)
        num3 = random.randint(10,50)
        max_records3 = num3
        l3 = mac_6_1[min_records3:(min_records3+max_records3)]#records to be published
       
        

    #Create the data object to publish with the selected rows
        print("MachineID: 6")
        print("No. of records", num3)
        print("Records from: ",min_records3,':',min_records3+max_records3)
        publish_message(producer, topic, l3)
        min_records3 += num3
  
        if max_records3 >= len(mac_6):
            max_records3 = 0
            min_records3 = 0
        
        '''MACHINE NO.4'''
        for value in mac_7:

            time_stamp =  int(dt.datetime.now().timestamp())
            value['ts']=time_stamp

            mac_7_1.append(value)
         #A random function to get random number of rows between (10,50)
        num4 = random.randint(10,50)
        max_records4 = num4
        l4 = mac_7_1[min_records4:(min_records4 + max_records4)]#records to be published
       
    #Create the data object to publish with the selected rows
        print("MachineID: 7")
        print("No. of records", num4)
        print("Records from: ",min_records4,':',min_records4+max_records4)
        publish_message(producer, topic, l4)
        min_records4 += num4

        if max_records4 >= len(mac_7):
            max_records4 = 0
            min_records4 = 0
            
        '''MACHINE NO.5'''
        #include timestamp as a key value pair for each record
        for value in mac_8:

            time_stamp =  int(dt.datetime.now().timestamp())
            value['ts']=time_stamp

            mac_8_1.append(value)
         #A random function to get random number of rows between (10,50)
        num5 = random.randint(10,50)
        max_records5 = num5
        l5 = mac_8_1[min_records5:(min_records5+max_records5)]#records to be published
       
        
    #Create the data object to publish with the selected rows
        print("MachineID: 8")
        print("No. of records", num5)
        print("Records from: ",min_records5,':',min_records5+max_records5)
        publish_message(producer, topic, l5)
        min_records5 += num5
        
        
        
    #Sleep for 5 seconds
        sleep(5)

        if max_records5 >= len(mac_8):
            max_records5 = 0
            min_records5 = 0
```
